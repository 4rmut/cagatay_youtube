```
from PIL import ImageDraw, Image
import face_recognition
from IPython.display import display
```

```
imge = face_recognition.load_image_file("s-9d11365b3cfad6c5aec302e1928a98ee0fcc7bf7.webp")
yuz_konumlari = face_recognition.face_locations(imge, number_of_times_to_upsample=3, model="cnn")
```

```

imge_pil = Image.fromarray(imge)
draw = ImageDraw.Draw(imge_pil)

for yuz in yuz_konumlari: 
    (tavan, sag, taban, sol) = yuz
    
    draw.rectangle([sol, tavan, sag, taban],  outline=(0, 0, 255))
    
```
