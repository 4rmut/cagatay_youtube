```
import torch
```

```
#@title Fonksiyon tanımı y = f(x)
def f(in_tensor):
    coeffs = torch.tensor([0.8, -0.5, 0.7], dtype=torch.float).unsqueeze(1) 
    return in_tensor.mm(coeffs) -0.2
```

```
# Veri seti oluşturalım
x = torch.rand((100, 3), dtype=torch.float)
y = f(x)

x_train = x[0:80, :]
x_val = x[80:100, :]
y_train = y[0:80, :]
y_val = y[80:100, :]

```

**Sistemi oluşturalım**

```
class Fonksiyon(torch.nn.Module):
  def __init__(self):
    super(Fonksiyon, self).__init__()
    self.carpan_katsayilar = torch.nn.Parameter(torch.rand((3, 1), dtype=torch.float))
    self.toplam_katsayilar = torch.nn.Parameter(torch.rand((1, 1), dtype=torch.float))

  def forward(self, x):
    return x.mm(self.carpan_katsayilar) + self.toplam_katsayilar
```

```
model = Fonksiyon()
```

```
loss_fonksiyonu = torch.nn.MSELoss(reduction='sum')

optimizer = torch.optim.Adam(model.parameters(), lr=0.1)
```

**Eğitim Başlasın!**

```
for i in range(100):
  y_tahmin = model(x_train)

  loss = loss_fonksiyonu(y_tahmin, y_train)
  
  optimizer.zero_grad()

  loss.backward()

  optimizer.step()

  # validation

  with torch.no_grad():
    y_tahmin_val = model(x_val)

    loss_val = loss_fonksiyonu(y_tahmin_val, y_val)

  print('------------------------')
  print('Tekrar: {}, Loss: {}'.format(i, loss.item())) 
  print('Bulunan fonksiyon: y = {}x_1 + {}x_2 + {}x_3 + {}'.format(
    model.carpan_katsayilar[0].item(), 
    model.carpan_katsayilar[1].item(),
    model.carpan_katsayilar[2].item(),
    model.toplam_katsayilar[0].item()             
  ) )

  
```
